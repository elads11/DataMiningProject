#### Focus on Ranking variable using Regex Extraction
In this part, we focus on an intersting cloumn called 'Ranking'. Ranking consists of 'X' of 'Y' format. X represents the ranking, and Y represents the number of entities in that catagory or city. We decided to seperate those 2 elements into 2 different columns, and dismissing the 'of' word. the ranking of each restaurant would be 'individual_rank', and the number of entities in the catagory would be: 'Ranking_catagory'.
we used regular  expression to extract the first occurence of X number into the new column, and we used $ to extract the last occurence of Y to the second new column.
then we sorted the values so first would appear the highest ranks restaurants throughout USA.

```python
data=data.rename(columns = {'Total Review':'TotalReview'}) #a rename for later to avoid errors
data = data.loc[data.Ranking.notnull()]

data['individual_rank'] = data['Ranking'].str.extract('([0-9?,]+)')
pd.to_numeric(data['individual_rank'], errors = 'ignore')

data['Ranking_catagory'] = data['Ranking'].str.extract('([0-9?,]+$)')
pd.to_numeric(data['Ranking_catagory'], errors = 'ignore')
data= data.sort_values('individual_rank')
USdata = data.loc[data['Country'] == 'United States']
USdata.head(5)
```

### Some text processing from an interesting column 
The column Cuisine decribes the type of food and gives general descripition of the restaurant.
from this point we will try to process and analyze this column and see what kind of food is the most common among the restaurans in the USA.

here we printed the unique words of the column

here we grouped by the  Cuisine column and we can see that american african Greek Middle eastern Cafe has a lot of occurences.

here we downloaded wordcloud, and visualized  the Cuisine's most common words. we also used stopwords, and we iterated for each row, and converted each word to string instead of object type.

```python
from wordcloud import WordCloud, STOPWORDS
import PIL
import itertools

CuisineData = narrowedData[['Cuisine']].copy()
CuisineData = CuisineData.loc[CuisineData.Cuisine.notnull()]
comment_words = ' '
stopwords = set(STOPWORDS)
 
for index, row in CuisineData.iterrows():
    row = str(row)
    tokens = row.split()
    
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
         
    for words in tokens:
        comment_words = comment_words + words + ' ' 

wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
 
# plot the WordCloud image                       
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
